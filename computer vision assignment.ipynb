```python

# example of loading the inception v3 model
from keras.applications.inception_v3 import InceptionV3
# load model
model = InceptionV3()
# summarize the model
model.summary()
```

```python
import tensorflow as tf
```

```python
from PIL import Image
import PIL
import numpy as np
def load_image(image_path):
    img = Image.open( image_path )
    newImg = img.resize((299,299), PIL.Image.BILINEAR).convert("RGB")
    data = np.array( newImg.getdata() )
    return 2*( data.reshape( (newImg.size[0], newImg.size[1], 3) ).astype( np.float32 )/255 ) - 1
```

```python
img = load_image("dog.jpg")
img = np.expand_dims(img, axis=0)
```

```python
predict = model.predict(img)
print('Predicted:', decode_predictions(predict))
```

```python
model.compile()
```
